```python
import numpy as np 
import matplotlib.pyplot as plt
```

Commençons par une définition de l'oscillateur harmonique classique.

C'est un objet mathématique qui représente l'évolution temporelle d'une grandeur d'un système physique, tel que la position ou l'impulsion, piégée dans un puit de potentiel harmonique. Ce potentiel est défini par l'équation symétrique:
$a*{(x+b)}^2 +c$, où a est un facteur d'agrandissement, b un décalage sur l'axe des abscisses et c un décalage sur l'axe des ordonnnées. Le potentiel le plus connu est: $\frac{1}{2}k{x(t)}^2$, qui est relié à la loi de Hooke pour les ressorts, k étant la raideur du ressort.

```python
x = np.linspace(-2,2,100)
y = 0.5*x**2 # We take k = 1
plt.plot(x,y)
plt.title(r"Puit de potentiel $\frac{1}{2}k{x}^2$")
plt.xlabel(r"postion x")
plt.ylabel(r"Energie potentielle")
plt.show()
```

Pour résoudre un problème de mécanique classique, tel que déterminer l'équation à laquelle obéit un ressort, nous devons utiliser une loi d'évolution. Si le système est conservatif et donc que les forces de frottements sont négligées, nous avons l'énergie mécanique qui s'exprime comme: E(t) = K(t) + U(t),

K représente l'énergie cinétique, qui est $\frac{1}{2}m{v(t)}^2$ tandis que U représente l'énergie ponteitel.Dans notre exemple, $U(t) = \frac{1}{2}k{x(t)}^2$

En utilisant la conservation de l'énergie mécanique $\frac{dE}{dt} = 0$ nous sommes en mesure de trouver l'équation différentielle qui régit l'évolution de la variable x(t).

$\frac{dE}{dt} = m\dot{x}\ddot{x} + k\dot{x}{x} = 0$

Ce qui nous amène à la bien connue équation de l'oscillateur harmonique classique:

$\ddot{x} + \frac{k}{m}{x} = {0}$

Cette équation étant linéaire et à coefficients constant, nous pouvons la résoudre facilement en résolvant son équation caractéristique.

${r}^2 + \frac{k}{m} = 0$

Ce qui nous donne $ r = \pm i\sqrt{\frac{k}{m}}$ 

et donc x(t) = $Ae^{-i\sqrt{\frac{k}{m}}t} + Be^{-i\sqrt{\frac{k}{m}}t}$, qu'on peut aussi écrire en réel comme $D\cos(\sqrt{\frac{k}{m}}t + \phi)$

A,B,D et $\phi$ sont des constantes à déterminer grâce aux conditions initiales.

Prenons comme exemple D = 1 et $\phi$ = 0 , k = 1 et m = 1

```python
t = np.linspace(0,10,100)
D,m,k,phi = 1,1,1,0

x = D*np.cos(np.sqrt(k/m)*t + phi)

fig, ax = plt.subplots()

ax.plot(t,x, label = 'm ={} kg, k = {} N/m, $\phi$ = {} rad '.format(m,k,phi))
ax.legend(loc = "upper right")

plt.title(r"position de la masse attachée à un ressort cours du temps")
plt.xlabel(r"temps t")
plt.ylabel(r"position x")
plt.show()
```

Nous venons de résoudre le cas d'un oscillateur harmonique classique, non amorti et/ou non forcé, c'est la cas le plus simple. Nous resterons dans le cas parfait pour faire la comparaison avec l'oscillateur harmonique quantique qui va suivre.

Dès lors qu'on s'intéresse à des objets physiques qui sont très petits, tel que les particules et les atomes, il n'est plus posible d'utilser le formalisme classique, nous devons passer par la mécanique quantique. 

Un cas concret pour illuster la pertinence de vouloir un modèle pour un oscillateur quantique harmonique vient du fait que si l'on souhaite par exemple modéliser un atome proche de l'équilibre dans un puit de potentiel, alors nous pouvons approximer celui-ci par un potentiel harmonique au au niveau du point d'équilibre par un développement limité à l'ordre 2 de ce puit de potentiel. Ainsi il est très courant de se ramener à l'étude des oscillateurs harmoniques tant en mécanique classique que quantique.

Nous allons aborder le problème de deux points de vue. D'abord par un appproche ondulatoire et ensuite matricielle.

Nous allons donc devoir résoudre l'équation de Shrodïnger en 1D. Cette équation est un analogue de l'énergie mécanique pour les systèmes quantiques.

la forme générale de cette équations aux dérivées partielles est : $\frac{-\hbar^2}{2m}\Delta{\psi(\overrightarrow{r},t)} + V(\overrightarrow{r},t)\psi(\overrightarrow{r},t) = i\hbar\frac{\partial\psi}{\partial t}$

Dans le cas unidimensionnel celle-ci devient : $\frac{-\hbar^2}{2m}\frac{\partial^2\psi(x,t)}{\partial x^2} + V(x,t)\psi(x,t) = i\hbar\frac{\partial\psi(x,t)}{\partial t}$

Dans le cas de l'oscillateur harmonique quantique, le potentiel harmonique est le même qu'en classique c'est-à-dire $\frac{1}{2}k{x}^2$. Mais ici nous allons réecrire k différement, il vient de l'analyse dimensionnelle que $\frac{k}{m}$ est équivalent à une pulsation ${\omega}^2$, nous pouvons donc écrire que $ k = m{\omega}^2$, ce qui nous permet de faire appaitre la fréquence propre de notre système.

Finalement l'équation que nous devons résoudre est : $\frac{-\hbar^2}{2m}\frac{\partial^2\psi(x,t)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2\psi(x,t) = i\hbar\frac{\partial\psi(x,t)}{\partial t}$

On constate que nos variables sont séparables, nous pouvons donc chercher une fonction $\psi(x,t) = F(x)T(t)$
En réinjectant dans l'équation nous obtenons :

$\frac{-\hbar^2}{2m}\frac{\partial^2 F(x)T(t)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2\ F(x)T(t) = i\hbar\frac{\partial\ F(x)T(t)}{\partial t}$

ce qui nous amène à  $\frac{-\hbar^2}{2m}T(t)\frac{\partial^2 F(x)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2\ F(x)T(t) = i\hbar F(x)\frac{\partial\ T(t)}{\partial t}$

Et en divisant par F(x)T(t) des deux côtés nous obtenons : $\frac{-\hbar^2}{2m}\frac{1}{F(x)}\frac{\partial^2 F(x)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2 = i\hbar\frac{1}{T(t)}\frac{\partial\ T(t)}{\partial t}$ ce qui est une équation aux variables séparées, ce qui signifie que pour qu'elle soit vraie pour $\forall x $ et $ \forall t$ , il est nécessaire que :

$\frac{-\hbar^2}{2m}\frac{1}{F(x)}\frac{\partial^2 F(x)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2 = E$
$\\$
$i\hbar\frac{1}{T(t)}\frac{\partial\ T(t)}{\partial t} = E$ 

E étant dimensionnellement une énergie, et mathématiquement une constante de séparation.

Nous pouvons désormais nous ramner à la résolution de deux équations différentielles.

Commençons par la plus simple, celle qui régit T(t), la partie temporelle de $\psi(x,t)$

$i\hbar\frac{1}{T(t)}\frac{\partial\ T(t)}{\partial t} = E$ 
$\\$
$\frac{\partial\ T(t)}{\partial t} = -i\frac{E}{\hbar}T(t)$

$\frac{\partial\ T(t)}{\partial t} + i\frac{E}{\hbar}T(t) = 0 $

En résolvant l'équation caractéristique associée : $r +i\frac{E}{\hbar} = 0$

On trouve que T(t) = $Ae^{-i\frac{E}{\hbar}t}$

Et maintenant le coeur du problème, la résolution de l'équation de Shrodinger indépendante du temps pour la partie spatiale.

$\frac{-\hbar^2}{2m}\frac{1}{F(x)}\frac{\partial^2 F(x)}{\partial x^2} + \frac{1}{2}m{\omega}^2{x}^2 = E$
$\\$
$\frac{-\hbar^2}{m}\frac{1}{F(x)}\frac{\partial^2 F(x)}{\partial x^2} + m{\omega}^2{x}^2 = 2E$

Maintenant on va chercher à adimensionner l'équation pour ne pas trainer des paramètres telle que la masse ou la pulsation. Pour celà on va construire des variables $\tilde{x}$ et  $\tilde{E}$ adimensionnées à partir des paramèrtres de l'équation pour construire des grandeurs caractéristiques du problèmes.

x étant une longueur, nous avons besoins de construire la longeur carectéristique du problème $x_0$ afin de créer $\tilde{x}$.

Les dimensions de nos paramètres sont les suivantes :

$[h] = M{L}^2{T}^-1$
$\\$
$[m] = M$

$[\omega] = {T}^-1$

$[E] = M{L}^2{T}^-2$

Nous pouvons donc construire $x_0$ comme étant $\sqrt{\frac{\hbar}{m\omega}}$ et donc $\tilde{x} = x/x_0$

De même on construit l'énergie caaractéristique $\epsilon_0 = \hbar\omega$ et donc $\tilde{E} = \frac{E}{\hbar\omega}$

Donc nous avons : $ x = x_0\tilde{x}$ et $E = \epsilon_0\tilde{E}$

Il nous faut désormais trouver comment s'exprime $\frac{d^2}{x^2}$ par rapport à la variable tilde.

$\frac{d}{d\tilde{x}} = \frac{d}{dx}\frac{dx}{d\tilde{x}}$ et vu que $x = x_0\tilde{x}$ celà signfie que $\frac{d}{d\tilde{x}} = x_0\frac{d}{dx}$. Ce qui nous donne dans le sens inverse $\frac{d}{dx} = \frac{1}{x_0}\frac{d}{d\tilde{x}}$. Finalement nous obtenons $\frac{d^2}{dx^2} = \frac{1}{x_0^2}\frac{d^2}{d\tilde{x}^2}$

Nous pouvons désormais réinjecter dans notre équation différentielle.

$-\frac{\hbar^2}{m}\frac{1}{F(\tilde{x})}\frac{1}{x_0^2}\frac{d^2 F(\tilde{x})}{d\tilde{x}^2} + m{\omega}^2{x_0^2\tilde{x}^2} = 2\epsilon_0\tilde{E}$

En développant :

$-\frac{\hbar^2}{m}\frac{1}{F(\tilde{x})}\frac{m\omega}{\hbar}\frac{d^2 F(\tilde{x})}{d\tilde{x}^2} + m{\omega}^2{\frac{\hbar}{m\omega}\tilde{x}^2} = 2\epsilon_0\tilde{E}$

Ce qui amène à :

$-\hbar\omega\frac{1}{F(\tilde{x})}\frac{d^2 F(\tilde{x})}{d\tilde{x}^2} + \hbar\omega\tilde{x}^2 = 2\hbar\omega\tilde{E}$

Et en divisant par $\hbar\omega$ on tombe sur :

$-\frac{1}{F(\tilde{x})}\frac{d^2 F(\tilde{x})}{d\tilde{x}^2} + \tilde{x^2} = 2\tilde{E}$

Finalement en réarrangeant lest termes notre équation adimensionnée est: 

$\frac{d^2 F(\tilde{x})}{d\tilde{x^2}} + (2\tilde{E} -\tilde{x}^2) F(\tilde{x}) =0 $

Désormais nous procédons à une analyse asymptotique de cette équation lorsque $ \lim \tilde{x}\to\infty$. Du fait que $2\tilde{E}$ soit une constante, nous obtenons l'équation différentielle $\frac{d^2 F(\tilde{x})}{d\tilde{x^2}} -\tilde{x}^2 F(\tilde{x}) =0 $ pour  $ \lim \tilde{x}\to\infty$

Nous posons un ansatz pour la solution asymptotique : $F_{asympt}(\tilde{x}) = e^{\phi(\tilde{x})}$ que nous réinjectons dans l'équation. 

Nous obtenons ainsi une équation différentielle pour $\phi(\tilde{x})$ qui est :

$\frac{d^2\phi}{d\tilde{x}^2}e^{\phi(\tilde{x})} + \left( \frac{d\phi}{d\tilde{x}} \right)^2e^{\phi(\tilde{x})} = \tilde{x}^2e^{\phi(\tilde{x})}$

ce qui nous amène par simplification ( les exponentielles étant jamais nulle ) à : 

$\frac{d^2\phi}{d \tilde{x}^2}+ \left( \frac{d\phi}{d\tilde{x}} \right)^2 = \tilde{x}^2$

Nous utilisons désormais le fait que nous cherchons une fonction $\phi(\tilde{x})$ pour les $\lim \tilde{x}\to\infty$

Cette condition nous permet de négliger $\frac{d^2\phi}{d \tilde{x}^2}$ car dans cette situation si nous avons $ \left( \frac{d\phi}{d\tilde{x}} \right)^2 \propto \tilde{x}^{2} $ alors $\frac{d\phi}{d\tilde{x}} \propto\pm \tilde{x}$ et donc $\frac{d^2\phi}{d \tilde{x}^2} \propto\pm 1$.

Et ici notre condition  $ \lim \tilde{x}\to\infty$ nous permet de négliger ce terme constant matérialsé par la dérivée seconde de $\phi(\tilde{x})$

Nous nous retrouvons donc avec une équation différentielle facile à résoudre :

$\frac{d\phi}{d\tilde{x}} = \pm \tilde{x}$

Qui nous donne $\phi(\tilde{x}) = \frac{\pm\tilde{x}^2}{2}$ et donc  $F_{asympt}(\tilde{x}) = e^{\frac{\pm\tilde{x}^2}{2}}$

Physiquement la solution $e^{\frac{\tilde{x}^{2}}{2}}$ n'étant pas acceptable, car celle-ci diverge, il reste que $F_{asympt}(\tilde{x}) = e^{\frac{-\tilde{x}^2}{2}}$

Il nous reste maintenant à trouver une fonction $P(\tilde{x})$ tel que $F(\tilde{x}) = P(\tilde{x})F_{asympt}(\tilde{x})$
pour  inclure les variation de $\tilde{x})$ dans le cas non asymptotique.

Pourquoi $P(\tilde{x})$ ? 

Car pour que  $\lim \lim_{\tilde{x} \to\infty}F(\tilde{x}) \to F_{asympt}(\tilde{x}) $, il est nécessaire que $P(\tilde{x})$ soit un polynôme pour garantir la décroissance rapide et permettre $\lim \lim_{\tilde{x} \to\infty}F(\tilde{x}) \to F_{asympt}(\tilde{x})$ et permettre à $F(\tilde{x})$ d'être de carré sommable et normalisable afin de pouvoir avoir $\int_{-\infty}^{\infty} |{F(\tilde{x})}|^2 dx =1$.

Il faut désormais réinjecter cette expression de $F(\tilde{x})$ dans $\frac{d^2 F(\tilde{x})}{d\tilde{x^2}} + (2\tilde{E} -\tilde{x}^2) F(\tilde{x}) =0 $ afin de trouver l'équation différentielle de $P(\tilde{x})$ et donc son expression.

En procédant ainsi nous obtenons :

$\left( \frac{d^2P(\tilde{x})}{d\tilde{x}^2} -2\tilde{x}\frac{dP(\tilde{x})}{d\tilde{x}} + (x^2 - 1)P(\tilde{x}) \right)e^{\frac{-x^2}{2}} = (x^2 -2\tilde{E})P(\tilde{x})e^{\frac{-x^2}{2}}$

On peut simplifier les gaussiennes de part et d'autres de l'équation et la simplifier en developpant et ramenant les termes du même côté, nous obtenons ainsi l'équation différentielle de $P(\tilde{x})$ :

$ \frac{d^2P(\tilde{x})}{d\tilde{x}^2} -2\tilde{x}\frac{dP(\tilde{x})}{d\tilde{x}} + (2\tilde{E} - 1)P(\tilde{x}) = 0$

Nous reconnaissons une équation différentielle de Hermite de la forme : y(x)" - 2xy(x)' + 2ny(x) = 0

Par comparaisons nous avons donc $2\tilde{E} - 1 = 2n$ ce qui nous amène à $ \tilde{E}_n = n + \frac{1}{2}$

De plus cette équation est connue, elle admet deux formes de solutions indépendantes, la seule convergente étant les polynômes de Hermite $H_n(x)$ qui nécessite que $ n\in \mathbb{N}$, celà amène donc naturellement à la  quantification de l'énergie.

Nous avons donc trouvé la solution de notre équation adimensionnée que l'on rappel :

$\frac{d^2 F(\tilde{x})}{d\tilde{x^2}} + (2\tilde{E} -\tilde{x}^2) F(\tilde{x}) =0 $

Cette solution est donc $F_n(\tilde{x}) = H_n(\tilde{x})e^{\frac{-\tilde{x}^2}{2}}$

$\tilde{x} = x/x_0 $ avec $x_0 =  \sqrt{\frac{\hbar}{m\omega}}$

Nous avons donc trouver notre solution en fonction de x en faisant les remplacement :

$F_n\left( x\sqrt\frac{m\omega}{\hbar}\right) = H_n\left( x\sqrt\frac{m\omega}{\hbar}\right)e^{\frac{-m\omega x^2}{2\hbar}}$

Il nous reste encore à la normaliser.
